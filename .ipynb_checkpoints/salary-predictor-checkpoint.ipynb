```python
df_clean = df[df['Salary'].str.contains('Lacs PA',na = False)].copy()
```

```python
df_clean.head()
```

```python
import re
```

```python
def parse_salary(salary_str):
    numbers = re.findall(r'\d+\.?\d*',salary_str)

    if len(numbers)>=2:
        min_sal = float(numbers[0])
        max_sal = float(numbers[1])
        return (min_sal+max_sal)/2
    return None
```

```python
df_clean['avg_salary_lpa'] = df_clean['Salary'].apply(parse_salary)
```

```python
df_clean = df_clean.dropna(subset=['avg_salary_lpa'])
```

```python
df_clean.head()
```

```python
df_clean['Location'] = df_clean['Location'].str.split(',').str[0].str.strip()
```

```python
print(df_clean['Location'].value_counts().head(5))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
```

```python
top_5_cities = df_clean['Location'].value_counts().nlargest(5).index
df_plot = df_clean[df_clean['Location'].isin(top_5_cities)]
sns.boxplot(x='Location',y='avg_salary_lpa',data = df_plot,palette = "viridis")
plt.title ('Salary distribution across top 5 cities of INDIA',fontsize=15)
plt.xlabel('cities',fontsize =12)
plt.ylabel('average salary',fontsize =12)
plt.show()
```
